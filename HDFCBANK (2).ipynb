```python
df.rename(columns={'%Deliverble': 'Deliverble'}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Trades'].fillna(df['Trades'].mean(), inplace=True)
df['Deliverable Volume'].fillna(df['Deliverable Volume'].mean(), inplace=True)
df['Deliverble'].fillna(df['Deliverble'].mean(), inplace=True)
```

```python
df['%Deliverble']
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Close'], kde=True)
plt.title('Distribution of Closing Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()

```

```python
sns.boxplot(x=df['Volume'])
plt.title('Boxplot of Volume')
plt.xlabel('Volume')
plt.show()
```

```python
sns.scatterplot(x='Open', y='Close', data=df)
plt.title('Open vs Close Price')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.show()
```
